#### 1.数据初始特征查看，缺失值，最大值，最小值，众数，取值个数，分位数等等，结合变量指标实际情况检查数据特征是否合理，是否可用

```python
def edd(data,var,ty='numeric'):
    #print 'ok'
    edd=pd.DataFrame()
    edd['var'] = [var]
    edd['cnt'] = [data.shape[0]]
    edd['mis_rate'] = [data[var].isnull().sum()/float(data.shape[0])]
    edd['zero_rate'] = [(data[var]==0).sum()/float(data.shape[0])]
    edd['mode'] = [data[var].mode().tolist()]
    edd['mode_rate'] = [data[var].isin(data[var].mode().tolist()).sum()/float(data.shape[0])]
    edd['unique_cnt'] = [len(data[data[var].notnull()][var].unique())]

    if ty=='numeric':
        edd['max']=[data[var].max()]
        edd['min']=[data[var].min()]
        edd['mean'] = [data[var].mean()]
        edd['std'] = [data[var].std()]
        edd['p0.05'] = [data[~data[var].isnull()][var].quantile(0.05)]
        edd['p0.1'] = [data[~data[var].isnull()][var].quantile(0.1)]
        edd['p0.2'] = [data[~data[var].isnull()][var].quantile(0.2)]
        edd['p0.3'] = [data[~data[var].isnull()][var].quantile(0.3)]
        edd['p0.4'] = [data[~data[var].isnull()][var].quantile(0.4)]
        edd['p0.5'] = [data[~data[var].isnull()][var].quantile(0.5)]
        edd['p0.6'] = [data[~data[var].isnull()][var].quantile(0.6)]
        edd['p0.7'] = [data[~data[var].isnull()][var].quantile(0.7)]
        edd['p0.8'] = [data[~data[var].isnull()][var].quantile(0.8)]
        edd['p0.9'] = [data[~data[var].isnull()][var].quantile(0.9)]
        edd['p0.95'] = [data[~data[var].isnull()][var].quantile(0.95)]
        edd['p0.99'] = [data[~data[var].isnull()][var].quantile(0.99)]
        
    return edd
    
def edd_vars(data,var):
    df = pd.DataFrame()
    for i,j in enumerate(data.dtypes):# 查看数据类型
        print i
        if data.columns[i] in var:
            if j=='object':
                ty='object'
                #print v
                ds=edd(data,data.columns[i],ty=ty)
            else:
                #print v
                ty='numeric'
                ds=edd(data,data.columns[i],ty=ty)
            df = pd.concat([df, ds])
        else:
            pass
    df = df.reset_index()
    df = df[['var','cnt','mis_rate','zero_rate','mode','mode_rate','unique_cnt','max','min','mean','std','p0.05','p0.1','p0.2','p0.3','p0.4','p0.5',
             'p0.6','p0.7','p0.8','p0.9','p0.95','p0.99']]

    
    return df
 
 #use:data_edd = edd_var(data,var)
```

```python
def bucket(df, var, max_bins=10):
    if (len(df[var].unique()) < max_bins) and (len(df[var].unique())>1):
        uvalue = np.sort(df[var].astype(float).unique())
        uvdiff = np.append(np.diff(uvalue).astype(float) / 2, 0)
        uvbucket = np.append(uvalue.min(), uvalue + uvdiff)
    elif len(df[var].unique()) == 1:
        print('only one vaule,can not bucket!')
        return False
    else:
        uvalue = np.empty(0)
        for i in np.arange(max_bins + 1):
            try:
                uvalue = np.unique((np.append(uvalue, df[var].quantile(float(i) / float(max_bins)))))
            except:
                pass
        uvdiff = np.append(np.diff(uvalue).astype(float) / 2, 0)
        uvbucket = np.append(uvalue.min(), uvalue + uvdiff)

    return uvbucket


def groupby_count(df, group, var, tp):
    if tp == 'numeric':
        uvbucket = bucket(df, var)
        if uvbucket is False:
            return False
        else:
            ds = pd.DataFrame()
            for g in df.groupby(group):
                df_g = g[1][[var]]
                df_g[var + '_bin'] = [tuple([float(j) for j in i.strip('([]').split(',')]) for i in
                                 np.array(pd.cut(df_g[var].astype(float), uvbucket, retbins=True, include_lowest=True)[0])]

                ds_t = pd.DataFrame(df_g[var + '_bin'].value_counts()).sort_index().reset_index()

                ds_t.rename(columns={'index': 'bins',
                                var + '_bin': str(g[0])+'_count'}, inplace=True)

                ds_t[str(g[0])+'_pert'] = ds_t[str(g[0])+'_count']/ds_t[str(g[0])+'_count'].sum()

                if ds.empty:
                    ds = ds_t
                else:
                    ds = pd.merge(ds, ds_t, how='inner', on='bins')

    elif tp == 'nominal':
        ds = pd.DataFrame()
        for g in df.groupby(group):
            df_g = g[1][var]
            ds_t = pd.DataFrame(df_g.value_counts()).sort_index().reset_index()
            ds_t.rename(columns={'index': 'value',
                                  var: str(g[0])+'_count'}, inplace=True)

            ds_t[str(g[0])+'_pert'] = ds_t[str(g[0])+'_count']/ds_t[str(g[0])+'_count'].sum()

            if ds.empty:
                ds = ds_t
            else:
                ds = pd.merge(ds, ds_t, how='inner', on='value')
    else:
        ds = pd.DataFrame()
        print('please input right parameter of tp: numeric or nominal, or you will get a null df!')

    return ds


# 计算的相对变化是最近两个时间的
def value_count_bytime(df, vars, group, tp):
    if tp == 'numeric':
        df_c = pd.DataFrame()
        t_s = df[group].unique().tolist()
        t_s.sort(reverse=True)
        t = [str(t) for t in t_s]
        date_num = len(t)
        for var in vars:
            ds = groupby_count(df, group, var, tp)
            if ds is not False:
                if date_num > 1:
                    ds['change'] = np.abs(ds[t[0]+'_pert'] - ds[t[1]+'_pert'])
                    ds['r_change'] = np.abs((ds[t[0]+'_pert'] - ds[t[1]+'_pert'])/ds[t[1]+'_pert'])
                else:
                    pass
                ds['var'] = var
                ds['left'] = ds.bins.apply(lambda x: round(x[0], 3))
                ds['right'] = ds.bins.apply(lambda x: round(x[1], 3))

                df_c = pd.concat([df_c, ds])
            else:
                pass

        df_c = df_c[['var', 'left', 'right'] + df_c.columns[0:-3].tolist()]
        del df_c['bins']
        df_c = df_c.reset_index()
        del df_c['index']

    elif tp == 'nominal':
        df_c = pd.DataFrame()
        t_s = df[group].unique().tolist()
        t_s.sort(reverse=True)
        t = [str(t) for t in t_s]
        date_num = len(t)
        for var in vars:
            ds = groupby_count(df, group, var, tp)
            if date_num > 1:
                ds['change'] = np.abs(ds[t[0] + '_pert'] - ds[t[1] + '_pert'])
                ds['r_change'] = np.abs((ds[t[0]+'_pert'] - ds[t[1]+'_pert'])/ds[t[1]+'_pert'])
            else:
                pass
            ds['var'] = var
            df_c = pd.concat([df_c, ds])

        df_c = df_c[['var'] + df_c.columns[0:-1].tolist()]
        df_c = df_c.reset_index()
        del df_c['index']
    else:
        df_c = pd.DataFrame()
        print('please input right parameter of tp: numeric or nominal, or you will get a null df!')

    return df_c
```
